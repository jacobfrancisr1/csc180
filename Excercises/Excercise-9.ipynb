# Exercise 9 – Insurance Purchase Prediction
- Predicts if a person buys insurance (yes/no).  
- Logistic regression for binary output.  
- predict_proba gives probability of buying.

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# Load dataset
df = pd.read_csv("/Users/jacobfrancis/dev/csc180/csv/insurance_data.csv")
df.head()
```

```python
# Visualize data
plt.scatter(df.age, df.bought_insurance, marker='+', color='red')
plt.xlabel('Age')
plt.ylabel('Bought Insurance')
plt.title('Age vs Insurance Purchase')
plt.show()
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[['age']], df.bought_insurance, test_size=0.1, random_state=42)
```

```python
# Train logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Evaluate model
print("Model Score:", model.score(X_test, y_test))
```

```python
# Predict probabilities and sample prediction
print("Predicted probabilities:\n", model.predict_proba(X_test))
print("Prediction for age 56:", model.predict([[56]])[0])
print("Prediction for age 25:", model.predict([[25]])[0])
```
