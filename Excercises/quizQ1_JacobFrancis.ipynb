## **Step 1: Optimal K**

test K = 1, 2, 3.  
K=1 has high error, K=3 is overfitting.  
So K=2 looks best (the “elbow” point).

## **Step 2: Find Centroids**

Start with:
C₁ = (25,3)  
C₂ = (40,7)

Point dist(C₁) dist(C₂) Cluster
- (25,3) | 0 | 15.52 | C₁ |
- (40,7) | 15.52 | 0 | C₂ |
-  (30,5) | 5.39 | 10.20 | C₁ |

Cluster 1 → P₁, P₃  
Cluster 2 → P₂  

New centroids:
C₁' = (27.5, 4)  
C₂' = (40, 7)


## **Step 3: Find WCSS**

Cluster 1 (27.5,4):  
(25,3) → (2.5² + 1²) = 7.25  
(30,5) → (2.5² + 1²) = 7.25  
Sum = 14.5  

Cluster 2 (40,7):  
(40,7) → 0  

## **Total WCSS = 14.5**


```python
# K-Means Quiz
# Jacob Francis - 11/04/2025

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# data points (Age, Miles per day)
X = np.array([[25,3],[40,7],[30,5]])

# manual centroids from calculation
manual_centroids = np.array([
    [27.5,4],  # cluster 1 center
    [40,7]     # cluster 2 center
])

# plot manual clusters and centroids
plt.scatter(X[:,0], X[:,1], color='blue', label='Data Points')
plt.scatter(manual_centroids[0,0], manual_centroids[0,1], color='red', marker='x', s=120, label='Centroid 1')
plt.scatter(manual_centroids[1,0], manual_centroids[1,1], color='green', marker='x', s=120, label='Centroid 2')
plt.xlabel('Age')
plt.ylabel('Miles per Day')
plt.title('Manual K-Means (K=2)')
plt.legend()
plt.show()

# run sklearn KMeans just to check my math
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)

# show centroids and WCSS (inertia)
print("Sklearn Centroids:\n", kmeans.cluster_centers_)
print("Sklearn Inertia (WCSS):", kmeans.inertia_)

# note: inertia is the same as WCSS, lower = tighter clusters
```

### Comparison

| Method | Centroids | WCSS |
|:--------|:-----------|:------|
| Manual | (27.5, 4), (40, 7) | 14.5 |
| sklearn | (27.5, 4), (40, 7) | 14.5 |

They match up closely, math checks out.

**Final:**
- K = 2  
- Centroids = (27.5, 4), (40, 7)  
- WCSS = 14.5
