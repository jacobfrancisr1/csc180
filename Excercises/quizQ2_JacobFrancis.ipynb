### Find the optimal K using sklearn.cluster.KMeans and draw the elbow plot

Dataset columns:
Country,Age,Height_cm,Weight_kg,Sex,Weight_in_Words

```python
# K-Means Elbow Method
# Jacob Francis - 11/04/2025

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# load csv file (update path if needed)
data = pd.read_csv("/Users/jacobfrancis/dev/csc180/csv/Data06JoMix.csv")

# look at first few rows
print(data.head())
```

```python
# use numeric columns for clustering
X = data[['Age', 'Height_cm', 'Weight_kg']]

# run KMeans for a few different K values
wcss = []  # within-cluster sum of squares
K_values = range(1, 10)  # range

for k in K_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)  # inertia = WCSS
```

```python
# plot the elbow
plt.figure(figsize=(6,5))
plt.plot(K_values, wcss, 'o-', color='blue')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS (Inertia)')
plt.title('Elbow Method for Optimal K')
plt.grid(True)
plt.show()
```

```python
# print out WCSS values to see how they drop
for k, val in zip(K_values, wcss):
    print(f"K={k}  -->  WCSS={val:.2f}")
```

### Observations

The "elbow" is where the WCSS starts to flatten out.  
From the plot, you can see the biggest drop happens before the curve levels — that’s your **optimal K**.

With this dataset, the elbow appears around **K = 2 or 3**
